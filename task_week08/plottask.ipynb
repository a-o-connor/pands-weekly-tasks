A program that displays:
- A histogram of a normal distribution of a 1000 values with a mean of 5 and standard deviation of 2
- A plot of the function  $h(x)=x^3$ in the range 0 to 10, on the one set of axes.
******

### Import Libraries
********

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Pseudo random number generator
**********
For random sampling, initialise a PNRG\
\
**References:**\
[Numpy Documentation on the Random Generator](https://numpy.org/doc/stable/reference/random/generator.html)\
[Real Python Tutorial on the Numpy PNRG](https://realpython.com/numpy-random-number-generator/)
******

```python
#Initialise random number generator 
rng = np.random.default_rng() #call rng to generate 
# No seed, so the numbers generated will be different each time
```

```python
#Generate a normal distribution of 1000 values with a mean of 5 and an SD of 2:

#Want to draw samples from a normal distribution
numbers = rng.normal(
    loc = 5,            #The mean
    scale = 2,          #The SD
    size = 1000         
)
#This returns a np array with a normal distribution
#rng initialise does not have a seed, so this will be different each time the cell is run
```

## Plot the Data
******

```python
# A quick histogram plot to have a look at the data:
plt.hist(numbers, edgecolor = "black")

plt.title("A Normal Distribution of 1000 Random Numbers")
plt.xlabel("Number")
plt.ylabel("Frequency")
```

```python
#Plot the function h(x) = x3 in the range 0-10

#Generate a numpy array for the x values 0 - 10
x_values = np.array(range(0,10))
#Cube the numpy array to generate the y values
y_values = x_values**3

# .plot() generates a simple line plot to have a look at the data 
plt.plot(
    x_values,
    y_values
)

#Some very unimaginative axis labels and chart title for a quick overview
plt.xlabel("Number")
plt.ylabel("Number Cubed")
plt.title("A plot of the function $h(x) = x3$")
```

### Plotting both on the same plot
And make it look good! 

**References:**\
[Real Python: Guide to Python Plotting With Matplotlib](https://realpython.com/python-matplotlib-guide/)\
\
(I'll be using the object oriented approach here as it was recommended in our Principles of Data Analytics lectures to get familiar with this approach.)
*********



```python
# subplot().  creates two variables, a matplotlib.figure and a matplotlib.axes
fig, ax = plt.subplots() #Defined one Figure (fig) containing one Axes (a subplot object, ax)

ax.hist(numbers, edgecolor = "black") #Call the ax method directly to create a histogram

#Call the ax method to set the axis labels and plot title
ax.set_xlabel("Number")
ax.set_ylabel("Frequency")
ax.set_title("What should the title of the plot be?")

# create a second axes for the h(x) = x3 function that shares the same x-axis
ax2 = ax.twinx() #ax2 will share the same y-axis as ax, through using the twinx function. 
ax2.plot(x_values, y_values, color = "red")
ax2.set_ylabel("Number Cubed")

```
